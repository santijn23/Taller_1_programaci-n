```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jun 18 02:48:50 2019

@author: Santiago Jiménez Naranjo
"""
#Ejercicio_12
## Este código lee dos palabras (en mayúscula o minúscula) y determina cual está
# primero en el diccionario. El programa soporta letras con tilde y la ü.

def convertir_a_minuscula(cadena):
    # Esta función convierte a minúscula la palabra ingresada utilizando el
    # comando .lower, de manera que se compare con mayor facilidad y se eviten 
    # ambigüedades y errores al comparar mayúsculas y minúsculas.
    
    cadena_minuscula= cadena.lower()
    return cadena_minuscula

def verificar_una_palabra(cadena):
    # Esta función verifica que se haya ingresado una sola palabra, asumiendo
    # que las palabras se separan con un carácter espacio ' '; la función eva-
    # lúa posición a posición y, en el momento que encuentra un espacio corta
    # la cadena y se evalúa la palabra inicial.
    
    for i in range (len(cadena)):
        if  cadena[i] == ' ':
            cadena_1= cadena[i - 1]
        else: cadena_1= cadena
    return cadena_1

def comparar_palabras(cad_1,cad_2,long):
    # Esta función compara dos palabras carácter a carácter, de manera que se
    # defina que palabra va primero en el diccionario. Evalúa si el carácter de
    # una posición específica es igual, luego se establecen una serie de condi- 
    # ciones en caso de que se encuentre en la misma posición una vocal con  
    # tilde y sin ella, ya que en el diccionario se siguen comparando los si-
    # guientes caracteres para definir cuál va primero; la tilde solo define en
    # el caso de que sean exactamente iguales, donde la palabra sin tilde va   
    # primero. Se añaden dos líneas más para ubicar la ñ entre la n y la o; al 
    # final, la función devuelve la palabra que va primero en el diccionario. 

    
    for n in range (long + 1):
        if n == long : break
    
        if ord(cad_1[n]) == ord(cad_2[n]):
            continue
        elif ord(cad_1[n]) == 225 and ord(cad_2[n]) == 97:  # á = 225, a = 97
            continue
        elif ord(cad_1[n]) == 97 and ord(cad_2[n]) == 225:
            continue
        elif ord(cad_1[n]) == 233 and ord(cad_2[n]) == 101: # é = 233, e = 101
            continue
        elif ord(cad_1[n]) == 101 and ord(cad_2[n]) == 233:
            continue
        elif ord(cad_1[n]) == 237 and ord(cad_2[n]) == 105: # í = 237, i = 105
            continue
        elif ord(cad_1[n]) == 105 and ord(cad_2[n]) == 237:
            continue
        elif ord(cad_1[n]) == 243 and ord(cad_2[n]) == 111: # ó = 243, o = 111
            continue
        elif ord(cad_1[n]) == 111 and ord(cad_2[n]) == 243:
            continue
        elif ord(cad_1[n]) == 250 and ord(cad_2[n]) == 117: # ú = 250, u = 117
            continue
        elif ord(cad_1[n]) == 117 and ord(cad_2[n]) == 250:
            continue
        elif ord(cad_1[n]) == 252 and ord(cad_2[n]) == 117: # ú = 252, u = 117
            continue
        elif ord(cad_1[n]) == 117 and ord(cad_2[n]) == 252:
            continue
        elif ord(cad_1[n]) == 241 and 111 <= ord(cad_2[n]) <= 122: # ñ = 241
             va_primero= cad_1                                     # o = 111
        elif ord(cad_2[n]) == 241 and 111 <= ord(cad_1[n]) <= 122: # z = 122
             va_primero= cad_2
        elif ord(cad_1[n]) > ord(cad_2[n]):
            va_primero= cad_2
        elif ord(cad_2[n]) > ord(cad_1[n]):
           va_primero = cad_1   
      
        return va_primero

# Se solicita al usuario ingresar dos palabras. 
        
while True :
    try:
        palabra_1= input('Por favor ingrese la primer palabra: ')
        break
    except :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

while True :
    try:
        palabra_2= input('Por favor ingrese la segunda palabra: ')
        break
    except :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')
        
# Se calcula la longitud de cada una de las palabras, ya que esto definirá el 
# rango en el que se evaluarán las demás funciones.
        
longitud_1= len(palabra_1)
longitud_2= len(palabra_2)

# Se ejecuta la función que evalúa si la cadena ingresada está compuesta por 
# una o más palabras.

verificar_1= verificar_una_palabra(palabra_1)
verificar_2= verificar_una_palabra(palabra_2)

# Se define el rango de evaluación entre las longitudes de la palabra 1 y 2, 
# eligiendo para este caso, la palabra que tenga mayor longitud.

if longitud_1 > longitud_2:
    longitud_mayor= longitud_1
else:
    longitud_mayor= longitud_2

# Se ejecuta la función para convertir a minúscula las palabras, de manera que 
# sea más fácil su comparación.
    
minuscula_1= convertir_a_minuscula(verificar_1)
minuscula_2= convertir_a_minuscula(verificar_2)

# Finalmente se ejecuta la función que compara ambas cadenas de texto en minús-
# cula y evalúa cual va primero.

va_primero= comparar_palabras(minuscula_1,minuscula_2,longitud_mayor)

# Se definen las condiciones de impresión para los 3 casos posibles, que las
# palabras sean iguales, en cuyo caso ocuparían el mismo lugar, que vaya prime-
# ro la palabra 1, o que vaya primero la palabra 2. 

if minuscula_1 == minuscula_2:
    print ('Las palabras ingresadas ocupan el mismo lugar en el diccionario.')
elif va_primero == minuscula_1:
    print (palabra_1,'va primero en el diccionario que', palabra_2)
else: # va_primero == minuscula_2:
    print (palabra_2,'va primero en el diccionario que', palabra_1 )
```

```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jun 17 23:08:50 2019

@author: Santiago Jiménez Naranjo
"""
#Ejercicio_09
#Este programa permite leer las coordenadas (x1, y1, r1) y (x2, y2, r2) que 
#corresponden al centro y al radio de dos círculos. Leer un punto de coordena-
#das (a, b). Y determina si (a, b) está contenido: a) dentro del círculo 1; b) 
#dentro del círculo 2; c) dentro de ambos círculos. Teniendo en cuenta que el 
#círculo es el lugar geométrico de los puntos del plano cuya distancia a otro 
#punto fijo, llamado centro, es menor o igual que una cantidad constante, 
#llamada radio.

def distancia(x, y, a, b):
#la función perimite calcular la ditancia e dos puntos, fundamentada por el
#teorema de pitágoras
    dist= (((x - a) ** 2)+((y - b) ** 2)) ** (1 / 2)
    return dist

# Ingresar los datos del círculo 1.
    
while True :
    try:
        coord_x_1= int(input('Ingresar la coordenada en X del círculo 1: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intentelo de nuevo')

while True :
    try:
        coord_y_1= int(input('Ingresar la coordenada en Y del círculo 1: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intentelo de nuevo')
        
#los ciclos while para las coordenadas del circulo 1 se usan para admitir uni-
#camente variables enteras, de manera que da la opción al usuario que si no lo
#hace, las ingrese nuevamente y de manera correcta.
        
while True :
    try:
        r_1= int(input('Ingrese el radio del círculo 1: '))
        if r_1 >= 0 : break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')
#El while para el radio tiene como misma función acpetar unicamente enteros

#datos para el circulo 2.
    
while True :
    try:
        coord_x_2= int(input('Ingresar la coordenada en X del círculo 2: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

while True :
    try:
        coord_y_2= int(input('Ingresar la coordenada en Y del círculo 2: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')
        
while True :
    try:
        r_2= int(input('Por favor ingrese el radio del círculo 2: '))
        if r_2 >= 0 : break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

# Ingresar las coordenadas del punto (a,b) a evaluar.
    
while True :
    try:
        a= int(input('Por favor ingrese la coordenada en X del punto: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

while True :
    try:
        b= int(input('Por favor ingrese la coordenada en Y del punto: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')

# Ejecutar la función nombrada al inicio de código.
        
distancia_a_1= distancia(coord_x_1, coord_y_1, a, b)
distancia_a_2= distancia(coord_x_2, coord_y_2, a, b)

# Evaluar las condiciones de impresión.

if distancia_a_1 <= r_1 and distancia_a_2 <= r_2:
    print ('\nEl punto', a,',', b, 'se encuentra dentro de ambos círculos.')
    
elif distancia_a_1 > r_1 and distancia_a_2 <= r_2:
    print ('\nEl punto', a,',', b, 'se encuentra solo dentro del círculo 2.')
    
elif distancia_a_1 <= r_1 and distancia_a_2 > r_2:
    print ('\nEl punto', a,',', b, 'se encuentra solo dentro del círculo 1.')
    
elif distancia_a_1 > r_1 and distancia_a_2 > r_2:
    print ('\nEl punto', a,',', b, 'no se encuentra dentro de ningún círculo.')
#los presentes condicionales if son para evalaur que las coordenadas ingresadas
#cumplas o no con las condiciones planteadas por el ejercicio
```

```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jun 17 23:40:20 2019

@author: Santiago Jiménez Naranjo
"""

#Ejercicio_07
#Leer un número entero positivo y determine si existen en el número dígitos
#repetidos. Reporte cuales son los dígitos repetidos. 

def contar_posiciones(num):
    #la presente función hace un conteo de cuántas posiciones tiene un número
    #mientras se tenga un número entero diferente de cero, se divide entre 10
    #y se suma 1 por cada división que sea posiblerealizar
    
    count=0
    while (num//10*10)>0:
        num=int(num//10)
        count += 1
    return count
def verificar_repetidos(num, n, dig):
    #Esta función permite verificar si se presentan dígitos repetidos en un 
    # número, comparando cada dígito con un número especificado.
    
    # El rango está dado para evaluar cada posición de izquierda a derecha, se 
    # extrae el dígito más a la izquierda, lo compara con un dígito en especí-
    # fico y se incrementa en 1 el contador por cada vez que se cumpla la con-
    # dición. Luego se resta al número inicial el producto de b por una poten-
    # cia de 10, de manera que se elimine el dígito más a la izquierda.
    
    count=0
    for i in range(-dig,1):
        b=num//(10**(-i))
        if b==n:
            count += 1
        num -= int(b*(10**(-i)))
    return count

# Se solicita al usuario ingresar un número entero positivo, en caso de que se  
# ingrese un valor negativo, un flotante o una letra, se solicita al usuario 
# ingresar el número otra vez.

while True:
    try:
        num= int(input('Ingrese el número entero positivo: '))
        if num > 0: break
    except ValueError:
        print('\nEl valor ingresado no es correct. Intentelo de nuevo: ')
        
#Se ejecuta la función definida al inicio
        
posiciones=contar_posiciones(num)
# A través de un ciclo for se verifican los dígitos repetidos entre 0 y 9 y se 
# imprimen los que se encuentran repetidos.
for n in range (0,10):
    repetido=verificar_repetidos(num,n,posiciones)
    if repetido > 1:
        print('el digito ',n, ' está repetido')
    else:
        print('El dígito ',n,'No está definido' )
```

```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jun 18 01:03:25 2019

@author: Santiago Jiménez Naranjo
"""
#Ejercicio_06
#Leer un número entero y determinar si la suma de sus dígitos es un número de
#Fibonacci. 

#definición de fibonacci: Se llaman números de Fibonacci a aquellos que forman 
#parte de la sucesión infinita de números naturales donde cada número se calcu-
#la sumando los dos anteriores a él.

def fibonacci(x):
    
    a=1
    b=1
    for i in range(x):
        s = a + b
        a = b
        b= s
        if s>= x: break
    return s

def conteo_posiciones(num):
    #esta función cuenta cuantas posiciones tiene el número
    #mientras haya un número positivo diferente de cero, se divide entr 10 y
    #se suma 1 po cada división que sea posible realizar.
    
    count=0
    while (num // 10 * 10) > 0 :
           num= int(num // 10)             
           count += 1
    return count

def sumar_digitos(num,count):
   # Esta función suma uno a uno los dígitos que componen un número.
    # El rango se define entre la cantidad de posiciones y 1, se extrae cada 
    # dígito de izquierda a derecha, se suma, y luego se resta al número ini-
    # cial el producto de dig por una potencia de 10, de manera que se elimine 
    # el dígito más a la izquierda.
    
    
    suma=0
    for n in range(-count,1):
        dig=int(num//(10**(-n)))
        suma += dig
        num -= int(dig*(10**(-n)))
    return suma

# Se solicita al usuario ingresar un número entero, en caso de que se  
# ingrese un flotante o una letra, se solicita al usuario ingresar el número
# nuevamente.
    
while True :
    try:
        num= int(input('Por favor ingrese un número entero: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')
#se ejecutan las funciones definidas anteriormente

posiciones= conteo_posiciones(abs(num))
suma= sumar_digitos(abs(num), posiciones)
es_fibonacci=fibonacci(suma)

#Se plantean las condiciones de impresión. Si el resultado de la serie de
#fibonacci es igual a la suma de los dígitos se concluye que es un número de
#fibonacci

if es_fibonacci==suma:
    print('la suma de los de dígitos del número ingresado es', suma, 'y es un '
          'número fibonacci')
else:
    print('\nla suma de los dígitos del número ingresado es',suma, 'y NO es un '
          'número fibonacci')
```

```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jun 15 01:52:35 2019

@author: Santiago Jiménez Naranjo
"""
#Ejercicio_05
#Leer un número entero y determinar el número de cincos que están consecutivos
#en un número.

def conteo_posiciones(num):
    
    count=0
    while (num//10*10)>0:
        num = int(num//10)
        count += 1
    return count
#el anterior condicional while permite fracmentar los números para que pedan
#ser evaluados

def conteo_cincos_conse(num,count):
    count_2_5= 0
    count_1_5= 0
    num_1= num
    x_1 = num // (10 ** (count))
    for n in range (-count, 1):
        num_1 -= x_1 * (10**(-n))
        x_2 = num_1 // (10**( -n - 1))
        if x_1 == x_2 == 5: #validando que el número es 5
            count_2_5 += 1
        elif x_2 == 5: #validando que el segundo número despues del primer 5 es 5
            count_1_5 += 1 
        x_1 = x_2
    return count_2_5, count_1_5

while True :  #validando que el dato que ingrese el ususario sea un entero
    try:
        num= int(input('Por favor ingrese un número entero: '))
        break
    except ValueError :
        print ('\nEl valor ingresado no es correcto, intente de nuevo')
        
num = abs(num)
count= conteo_posiciones(num)
cincos_consecutivos= conteo_cincos_conse(num, count)

if cincos_consecutivos[0] == 0:
    print ('No hay cincos consecutivos')
elif cincos_consecutivos[1] == 0:
    print ('Se tiene', cincos_consecutivos[0]+1, ' pareja de cincos consecuti'
           'vos')
else:
    print ('Se tiene', cincos_consecutivos[1], 'parejas de cincos consecutivos')
```
